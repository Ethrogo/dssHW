```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('Top_Highest_Openings.csv')

```

```python
top_10_movies = df.nlargest(10, 'Total Gross')
plt.figure(figsize=(10, 6))
plt.barh(top_10_movies['Release'], top_10_movies['Total Gross'], color='skyblue')
plt.xlabel('Total Gross ($)')
plt.ylabel('Movie')
plt.title('Top 10 Movies by Total Gross Earnings')
plt.gca().invert_yaxis()  # Invert y-axis to display the highest grossing movie at the top
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Opening'], df['Total Gross'], alpha=0.5, color='orange')
plt.xlabel('Opening Weekend Earnings ($)')
plt.ylabel('Total Gross Earnings ($)')
plt.title('Opening Weekend Earnings vs Total Gross Earnings')
plt.grid(True)
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])  # Convert 'Date' column to datetime format
df_sorted_by_date = df.sort_values('Date')  # Sort DataFrame by date
plt.figure(figsize=(10, 6))
plt.plot(df_sorted_by_date['Date'], df_sorted_by_date['Average'], marker='o', color='green', linestyle='-')
plt.xlabel('Date')
plt.ylabel('Average Earnings per Theater ($)')
plt.title('Trend of Average Earnings per Theater Over Time')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()
```
